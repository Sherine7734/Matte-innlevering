## **Nedkjøling av vann**

### I denne oppgaven målte jeg temperaturen til 78°C varmt vann i en kvadratisk ildfast form i glass i et rom med temperatur lik 25°C

### **Spesifikasjoner**
Vannets temperatur ved start: T<sub>0</sub> = 78°C\
Romtemperatur: T<sub>K</sub> = 24°C\
Vann volum: 5 dl\
Overflate areal: 15 cm x 15 cm = 225 cm<sup>2</sup>



```python
import numpy as np 
import matplotlib.pyplot as plt 
from math import e

t_målt = np.array([0, 80, 115, 140, 240, 285, 375, 525, 660, 760, 885, 1030, 1180, 1360, 1605, 2270, 3140, 3555, 4050, 4800, 5670, 6430, 7140, 7860]) 
temp_målt = np.array([78, 74, 72, 70, 68, 66, 64, 60, 58, 56, 54, 52, 50, 48, 45, 40, 36, 34, 32, 30, 28, 27, 26, 25])

t_verdier = np.linspace(0, 9000, 10000)

def T(t):
    return 25 + 53*e**(-0.0009809*t)

def T_2(t):
    return 25 + 53*e**(-0.0006814*t)

def T_3(t):
    return 25 + 53*e**(-0.0005065*t)

plt.plot(t_verdier, T(t_verdier), "g", label="Newtons avkjølingslov, T(80) = 74")
plt.plot(t_verdier, T_2(t_verdier), "c", label="Newtons avkjølingslov, T(885) = 54")
plt.plot(t_verdier, T_3(t_verdier), "b", label="Newtons avkjølingslov, T(5670) = 28")
plt.plot(t_målt, temp_målt, "r", label="Målt temperatur")

plt.xlabel("t [s]")
plt.ylabel(f"T [{chr(176)}C]")
plt.title(f"Avkjøling av vann på 78{chr(176)}C i rom på 25{chr(176)}C\n\n Figur 1\n")

plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from math import e

t_målt = np.array([0, 80, 115, 140, 240, 285, 375, 525, 660, 760, 885, 1030, 1180, 1360, 1605, 2270, 3140, 3555, 4050, 4800, 5670, 6430, 7140, 7860]) 
temp_målt = np.array([78, 74, 72, 70, 68, 66, 64, 60, 58, 56, 54, 52, 50, 48, 45, 40, 36, 34, 32, 30, 28, 27, 26, 25])

t_verdier = np.linspace(0, 9000, 10000)


def T_3(t):
    return 25 + 53*e**(-0.0005065*t)

plt.plot(t_verdier, T_3(t_verdier), "b", label="Newtons avkjølingslov, T(5670) = 28")
plt.plot(t_målt, temp_målt, "r", label="Målt temperatur")

plt.xlabel("t [s]")
plt.ylabel(f"T [{chr(176)}C]")
plt.title(f"Avkjøling av vann på 78{chr(176)}C i rom på 25{chr(176)}C\n\n Figur 2\n")

plt.grid()
plt.legend()
plt.show()
```

### **Newtons Avkjølingslov**

 $\dot T$(t) = 	$\alpha$(T(t) - T<sub>K</sub>)\
 T(0) = T<sub>0</sub>

 T<sub>0</sub> = 78°C\
 T<sub>K</sub> = 24°C


 Ut fra initialkravet og verdiene ovenfor fikk jeg følgende uttrykk for temperaturen:
 
T = T<sub>K</sub> + C e<sup>αt</sup>\
T = 25 + 53e<sup>αt</sup>

Jeg regnet ut tre ulike verdier av $\alpha$ ved å ta utgangspunkt i ulike tidspunkt fra målingen og plottet disse sammen med kurven for målingene. Se figur 1. 

Med hensyn på T(80) = 74\
$\alpha$<sub>1</sub> = ln(49/53)/80 ≈ -0.0009809

Med hensyn på T(885) = 54\
$\alpha$<sub>2</sub> = ln(29/53)/885 ≈ -0.0006814

Med hensyn på T(5670) = 28\
$\alpha$<sub>3</sub> = ln(3/53)/5670 ≈ -0.0005065

Vi ser at Newtons avkjølingslov for $\alpha$<sub>1</sub> og $\alpha$<sub>2</sub> leder til en raskere avkjøling enn det vi får målt i virkeligheten. Dette er fordi det ekstra kravet vi setter er fra en måling relativ kort tid etter start, og hvis man da ser på endringen i temperatur ut ifra initialkravet og det nye kravet vil man tro at vannet kjøles ned mye raskere enn det egentlig gjør, og vi får en større negativ stigning på grafen fra Newtons avkjølingslov. 

Vi ser allikevel at verdien for $\alpha$<sub>2</sub> med hensyn på T(885) = 54, altså en verdi som er målt lenger tid etter start, gir et uttrykk for T som er mer likt målingene enn verdien som ble beregnet ut ifra T(80) = 74 og $\alpha$<sub>1</sub>. 

For $\alpha$<sub>3</sub> er kravet satt mot slutten av målingene der verdien nærmer seg romtemperatur, og vil dermed gi en mest mulig nøyaktig verdi av $\alpha$ for vårt system. Se figur 2 for funksjonen fra $\alpha$<sub>3</sub> plottet sammen med målingene. Vi ser nå at temperaturen fra denne funksjonen synker raskere enn målingene våre i starten, før målingene  etterhvert følger grafen fra Newtons avkjølingslov svært nøyaktig ned mot 25°C. Det at modellen gir en saktere temperaturendring i starten, kan være grunnet at Newtons avkjølingslov ikke tar høyde for fordampning. 

Ut ifra grafene i figur 2 og observasjonene vi har gjort, kan vi konkludere med at Newtons avkjølingslov gir en god modell for å forutsi og uttrykke nedkjøling av vann over tid.
